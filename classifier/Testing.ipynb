```python
import requests
import os
from collections import Counter
from ML.TextPreprocessing import preprocess
from ML.TelecomClassifier import pipeline
import numpy as np
```

```python
words_pos = Counter()
words_neg = Counter()
```

```python
def bigrams(text):
    prev = text[0]
    text_V2 = []
    for i in range(1,len(text)):
        new_text = prev + " " + text[i]
        text_V2.append(new_text)
        prev = text[i]
    return text_V2
```

```python
path = 'A:\\Users\Artem\PycharmProjects\kafkaTest\\test' #путь где лежат файлы с отзывами
textx = []
mapping = []
for directory in os.listdir(path):
    if os.path.isdir(path + '\\' +directory):
        files = np.array(os.listdir(path + '\\' + directory))
        for file in files:
            with open(os.path.join(path + '\\' + directory + '\\', file), encoding='windows-1251') as f:
                textx.append(f.read())
                mapping.append(directory)             
```

```python
count = 0
for i in range(len(textx)):
    body = {
    "text": textx[i]
    }
    req = requests.post(url, data=body)
    
    result = req.json()['result']
    if result == 'negative' and mapping[i] == "neg":
        print("OK")
        words_neg.update(bigrams(preprocess(textx[i])))
    elif result == 'positive' and mapping[i] == "pos":
        print("OK")
        words_pos.update(bigrams(preprocess(textx[i])))
    else:
        count += 1
```
